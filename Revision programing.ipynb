```python
def extended_gcd(a, b):
    if b == 0:
        return (1, 0)
    else:
        x1, y1 = extended_gcd(b, a % b)
        x, y = y1, x1 - (a // b) * y1
        return x, y

def CR(list1, list2):
    if len(list1) == 1:
        return list1[0] % list2[0]

    a1, a2 = list1[0], list1[1]
    p1, p2 = list2[0], list2[1]
    s, t = extended_gcd(p1, p2)
    x12 = (a2 * s * p1 + a1 * t * p2) % (p1 * p2)

    return CR([x12] + list1[2:], [p1 * p2] + list2[2:])
```

## factorial of a number

```python
def fac(n):
    prod=1
    if n==0:
        return 1
    elif n==1:
        return prod
    else:
        for n in range(2,n+1):
            prod=prod*n
        return prod
```

## Calculate the sum of digits in a number

```python
def sum_of_digits(n):
    """
    Calculates the sum of digits using arithmetic operations only.
    Works for both positive and negative numbers.
    """
    n = abs(n)      # handle negative numbers
    total = 0
    while n > 0:
        total += n % 10   # get the last digit
        n //= 10         # remove the last digit
    return total
```

```python
sum_of_digits(1234)
```

```python
def sum_of_digits(n):
    """
    Another simple approach using list comprehension.
    """
    return sum([int(d) for d in str(abs(n))])
```

## Check if a string is a palindrome

```python
def is_palindrome(s):
    """
    Checks whether a given string reads the same backward and forward.
    Ignores spaces and capitalization.
    """
    s = s.replace(" ", "").lower()
    if s == s[::-1]:
        return f"'{s}' is a palindrome."
    else:
        return f"'{s}' is not a palindrome."
```

```python
is_palindrome("madam")
```

## Calculate the power of a number

```python
def power(base, exp):
    """
    Computes base raised to the power of exp.
    Equivalent to base ** exp.
    """
    result = 1
    for _ in range(exp):
        result *= base
    return result
```

## Check if a year is a leap year

```python
def is_leap_year(year):
    """
    Determines if a given year is a leap year.
    Follows Gregorian calendar rules.
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return f"{year} is a leap year."
    else:
        return f"{year} is not a leap year."
```

```python
is_leap_year(2004)
```

```python
is_leap_year(1997)
```

## Check if a number is prime

```python
def is_prime(n):
    """
    Determines if a given number is prime.
    Prime numbers are greater than 1 and divisible only by 1 and themselves.
    """
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

## Print the Fibonacci sequence

```python
def fibonacci(n):
    """
    Generates the first n terms of the Fibonacci sequence.
    """
    a, b = 0, 1
    sequence = []
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence
```

## Swap two variables using a temporary variable

```python
def swap_temp(a, b):
    """
    Swaps two variables using a temporary third variable.
    """
    temp = a
    a = b
    b = temp
    return a, b
```

## Swap two variables without using a temporary variable

```python
def swap_no_temp(a, b):
    """
    Swaps two variables without using an extra variable.
    Uses tuple unpacking.
    """
    a, b = b, a
    return a, b
```

## Print the multiplication table of a number

```python
def multiplication_table(n):
    """
    Prints the multiplication table for a given number up to 10.
    """
    table = []
    for i in range(1, 11):
        line = f"{n} x {i} = {n * i}"
        table.append(line)
    return table
```

## Find the largest element in a 1D array

```python
def largest_in_array(arr):
    """
    Returns the largest element in a list (1D array).
    """
    max_val = arr[0]
    for val in arr:
        if val > max_val:
            max_val = val
    return max_val
```

## Find the smallest element in a 1D array

```python
def smallest_in_array(arr):
    """
    Returns the smallest element in a list.
    """
    min_val = arr[0]
    for val in arr:
        if val < min_val:
            min_val = val
    return min_val
```

## Find the transpose of a 2D array

```python
def transpose_matrix(matrix):
    """
    Returns the transpose of a given 2D matrix.
    """
    rows = len(matrix)
    cols = len(matrix[0])
    transpose = [[0 for _ in range(rows)] for _ in range(cols)]
    for i in range(rows):
        for j in range(cols):
            transpose[j][i] = matrix[i][j]
    return transpose
```
